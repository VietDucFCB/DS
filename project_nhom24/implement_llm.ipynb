Tuy nhiên, để cải thiện hơn nữa, có thể xem xét sử dụng các mô hình ngôn ngữ hiện đại hơn và tối ưu hóa quy trình xử lý dữ liệu. Các ứng dụng tiềm năng của mã nguồn này trong thực tế bao gồm phân tích thị trường bất động sản, dự đoán giá nhà, và hỗ trợ quyết định đầu tư.

Cách thức hoạt động:
1. Chuẩn bị môi trường và dữ liệu
    - Import thư viện: Mã nguồn sử dụng các thư viện cơ bản như os, time, requests, pandas, json, re, và concurrent.futures để thực hiện các chức năng như gửi yêu cầu API, xử lý dữ liệu và triển khai đa luồng.
    - API base URL và token: Lấy thông tin về base URL của API và token từ biến môi trường để truy cập Anyscale API.
    - Dữ liệu đầu vào: data_for_llm là một DataFrame đã được định nghĩa từ trước, chứa các cột 'Listing ID' và 'content_for_llm'. Các cột này đã được đổi tên thành 'Listing ID' và 'text' để phù hợp với yêu cầu của hàm extract_features.
2. Hàm extract_features
    - Mục đích: Hàm này nhận vào một đoạn văn bản mô tả bất động sản và gửi yêu cầu đến Anyscale API để trích xuất các đặc trưng như giá, diện tích, địa chỉ, số phòng ngủ, v.v.
    - Thiết lập yêu cầu API: Tạo một yêu cầu HTTP POST với prompt là nội dung mô tả các đặc trưng cần trích xuất từ văn bản, và body là các thông tin khác như model sử dụng, nhiệt độ (temperature mang đặc tính sáng tạo của mô hình, trong trường hợp này là 0.5 để cho mô hình không được sáng tạo quá gây ra hiện tượng hallucination) và số lượng token tối đa (max_tokens).
    - Gửi yêu cầu và xử lý kết quả: Hàm gửi yêu cầu đến Anyscale API thông qua session đã thiết lập trước đó. Nếu thành công, hàm nhận và xử lý kết quả trả về để trích xuất các đặc trưng từ JSON được trả về.
    - Xử lý lỗi và tái thử lại: Nếu gặp lỗi trong quá trình gửi yêu cầu hoặc xử lý kết quả, hàm sẽ thử lại tối đa retry lần trước khi bỏ qua.
3. Hàm process_batch và process_dataframe
    - Hàm process_batch: Xử lý một batch DataFrame bằng cách sử dụng ThreadPoolExecutor để triển khai đa luồng. Mỗi mẫu dữ liệu từ batch được chuyển đến hàm extract_features để trích xuất các đặc trưng.
    - Hàm process_dataframe: Chia dữ liệu thành các batch và xử lý từng batch bằng cách sử dụng hàm process_batch. Kết quả từ mỗi batch được lưu vào result_dfs và sau đó ghép lại thành một DataFrame lớn.
4. Lưu kết quả vào tệp CSV
    - Lưu kết quả: Kết quả cuối cùng từ process_dataframe được lưu vào một tệp CSV với tên raw_data_llm.csv. Nếu tệp chưa tồn tại, hàm sẽ tạo mới với header; nếu đã tồn tại, dữ liệu sẽ được ghi vào mà không có header.

Mã nguồn này cung cấp một cách tiếp cận tự động hóa để trích xuất các đặc trưng từ văn bản mô tả bất động sản bằng cách sử dụng Anyscale API và xử lý dữ liệu một cách hiệu quả thông qua đa luồng. Kết quả được lưu trữ dưới dạng tệp CSV để dễ dàng quản lý và phân tích sau này.

```python
import os
import time
import requests
import pandas as pd
import json
import re
from concurrent.futures import ThreadPoolExecutor, as_completed
from requests.exceptions import RequestException

# Lấy thông tin API base URL và token từ biến môi trường
api_base = os.getenv("ANYSCALE_API_BASE", "https://api.endpoints.anyscale.com/v1")
api_token = os.getenv("ANYSCALE_API_TOKEN", "esecret_rkmhcga2whfl1r9jt7b2shdh9c")


# Giả sử df đã được định nghĩa và chứa các cột 'Listing ID' và 'content_for_llm'
data_for_llm = df[['Listing ID', 'content_for_llm']].rename(columns={'content_for_llm': 'text'})

def extract_features(text, retry=3, session=None):
    """
    Trích xuất các đặc trưng từ văn bản mô tả bất động sản bằng Anyscale.
    Hàm này sẽ thử lại tối đa `retry` lần nếu gặp lỗi.
    """
    if session is None:
        session = requests.Session()  # Tạo session nếu chưa có
    
    url = f"{api_base}/chat/completions"
    
    for attempt in range(retry):
        try:
            prompt = f"""
            Extract the following features from the real estate listing text IN VIETNAMESE:
            - Giá (ví dụ: 1200 triệu, 1,2 tỷ VND)
            - Diện tích (ví dụ: 120 m2, 50m x 40m)
            - Địa chỉ (địa chỉ đầy đủ nếu có, nếu không thì vị trí chung)
            - Số phòng ngủ
            - Số phòng tắm
            - Số tầng (nếu có)
            - Tiện ích (danh sách các tiện ích, cách nhau bằng dấu phẩy)
            - Tên đường
            - Tên phường, xã
            - Tên quận, huyện
            - Số mặt tiền
            - Nhà trong hẻm hay ngoài lộ

            Nội dung: {text}

            Cung cấp các đặc trưng được trích xuất theo định dạng JSON sau:
            {{
                "Price": "Giá được trích xuất",
                "Area": "Diện tích được trích xuất",
                "Address": "Địa chỉ được trích xuất",
                "Number of bedrooms": "Số phòng ngủ được trích xuất",
                "Number of bathrooms": "Số phòng tắm được trích xuất",
                "Number of floors": "Số tầng được trích xuất",
                "Amenities": "Các tiện ích được trích xuất",
                "Street name": "Tên đường được trích xuất",
                "Ward name": "Tên phường, xã được trích xuất",
                "District name": "Tên quận, huyện được trích xuất",
                "Number of frontages": "Số mặt tiền được trích xuất",
                "Alley or main road": "Nhà trong hẻm hay ngoài lộ"
            }}
            """

            body = {
                "model": "mistralai/Mixtral-8x22B-Instruct-v0.1",  # Chọn model phù hợp trên Anyscale
                "messages": [
                    {"role": "system", "content": "You are a helpful assistant."},
                    {"role": "user", "content": prompt}
                ],
                "temperature": 0.5,
                "max_tokens": 512  # Tăng số lượng tokens tối đa để có đủ chỗ cho dữ liệu bổ sung
            }

            # Gửi yêu cầu đến Anyscale API
            response = session.post(url, headers={"Authorization": f"Bearer {api_token}"}, json=body)
            response.raise_for_status()

            # Trích xuất kết quả và xử lý lỗi
            result = response.json()
            result_text = result.get('choices', [{}])[0].get('message', {}).get('content', '')

            # In phản hồi từ API để kiểm tra
            print(f"API response for text: {text}\nResponse: {result_text}\n")

            # Sử dụng regex để tìm phần JSON hợp lệ
            match = re.search(r'\{.*\}', result_text, re.DOTALL)
            if match:
                json_str = match.group(0)
                features_dict = json.loads(json_str)
                print(f"Parsed JSON: {features_dict}")

            else:
                print(f"Model did not return valid JSON: {result_text}")
                features_dict = {
                    "Price": "N/A",
                    "Area": "N/A",
                    "Address": "N/A",
                    "Number of bedrooms": "N/A",
                    "Number of bathrooms": "N/A",
                    "Number of floors": "N/A",
                    "Amenities": "N/A",
                    "Street name": "N/A",
                    "Ward name": "N/A",
                    "District name": "N/A",
                    "Number of frontages": "N/A",
                    "Alley or main road": "N/A",
                }
            return features_dict
        except (RequestException, json.JSONDecodeError) as e:
            print(f"Error: {e}, retrying in 1 second...")  # Print error for debugging
            time.sleep(1)
            if attempt == retry - 1:
                print(f"Failed after {retry} attempts: {e}")
                return {
                    "Price": "N/A",
                    "Area": "N/A",
                    "Address": "N/A",
                    "Number of bedrooms": "N/A",
                    "Number of bathrooms": "N/A",
                    "Number of floors": "N/A",
                    "Amenities": "N/A",
                    "Street name": "N/A",
                    "Ward name": "N/A",
                    "District name": "N/A",
                    "Number of frontages": "N/A",
                    "Alley or main road": "N/A",
                }

def process_batch(df_batch, session):
    """
    Xử lý một batch DataFrame bằng cách triển khai đa luồng
    """
    futures = []
    with ThreadPoolExecutor(max_workers=6) as executor:
        for _, row in df_batch.iterrows():
            text = row['text']
            futures.append(executor.submit(extract_features, text, session=session))
    
    results = []
    for future in as_completed(futures):
        features = future.result()
        results.append(features)
    
    result_df_batch = pd.DataFrame(results)
    result_df_batch['Listing ID'] = df_batch['Listing ID'].values
    return result_df_batch

def process_dataframe(df, batch_size=10):
    with requests.Session() as s:
        num_batches = len(df) // batch_size + (1 if len(df) % batch_size != 0 else 0)
        result_dfs = []
        for i in range(num_batches):
            batch_start = i * batch_size
            batch_end = min((i + 1) * batch_size, len(df))
            df_batch = df.iloc[batch_start:batch_end]
            result_df_batch = process_batch(df_batch, session=s)
            result_dfs.append(result_df_batch)
            print(f"Processed batch {i + 1}/{num_batches}")

    return pd.concat(result_dfs, ignore_index=True)
    

# Lấy mẫu dữ liệu từ dòng a đến dòng b
sample_data = data_for_llm.iloc[9500:9834]
# Xử lý DataFrame mẫu
result_sample_df = process_dataframe(sample_data, batch_size=10)

# Đường dẫn đến tệp CSV đầu ra
output_file = 'raw_data_llm.csv'

# Kiểm tra xem tệp đã tồn tại chưa
if not os.path.isfile(output_file):
    # Nếu chưa tồn tại, ghi DataFrame với header
    result_sample_df.to_csv(output_file, index=False, encoding='utf-8-sig')
else:
    # Nếu đã tồn tại, ghi DataFrame mà không có header
    result_sample_df.to_csv(output_file, index=False, encoding='utf-8-sig', mode='a', header=False)

print(f"Kết quả đã được lưu vào tệp {output_file}")
result_sample_df
```
